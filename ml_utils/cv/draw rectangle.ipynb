```python
%matplotlib inline

from typing import Tuple

import matplotlib.pyplot as plt
import numpy as np

from utils import draw_image_with_rectangle, Rectangle, iou
```

```python
def generate_image(height: int, width: int) -> Tuple[np.ndarray, Rectangle]:
    image = np.zeros((height, width, 3))
    
    x0 = np.random.randint(int(0.9 * height))
    y0 = np.random.randint(int(0.9 * width))
    x1 = np.random.randint(x0, int(0.99 * height))
    y1 = np.random.randint(y0, int(0.99 * width))
    
    image[x0:x1, y0:y1] = 1
    return image, Rectangle(x=y0 / width, y=x0 / height, width=(y1 - y0) / width, height=(x1 - x0) / height)
```

```python
image, r = generate_image(width=100, height=50)
```

```python
draw_image_with_rectangle(image, [r])
```
