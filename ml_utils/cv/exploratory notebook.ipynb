### Walidacyjny zbir zostanie zbudowany. Z folderu treningowego z kadej klasy zostanie przeniesionych 200 obrazkw.

```python
for class_ in train_distribution:
    class_path = os.path.join(TRAIN_DIRECTORY, class_)
    images_paths = os.listdir(class_path)
    valid_image_names = np.random.choice(images_paths, size=200, replace=False)
    
    for image_name in valid_image_names:
        move_object(os.path.join(class_path, image_name), os.path.join('valid', class_))
```

```python
VALID_DIRECTORY = 'valid'
```

```python
valid_distribution = get_distribution(VALID_DIRECTORY)
```
